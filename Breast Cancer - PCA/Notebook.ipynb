# Principal Component Analysis

PCA is an unsupervised Learning. It is used to examine the interrelations between the variables in order to identify the underlying structure of these variable .
Also know as general Factor Analysis.

Components are the transformation that chooses the variable system for the dataset such that the Greatest Variance of the dataset comes to lie at the First Axis , Second Greatest Variance at Second Axis and so on.

Components chooses the variables which are explaining the most (means which are giving more information or say useful ).

First line explains 70% of the variation. Second line explains 28% of the variation and 2% of the variation remains unexplained.

If we use the technique of PCA on a dataset with large number of variables, we can compress the amount of explained variation to just a few components.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
df = pd.DataFrame(cancer['data'], columns =cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
# PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,8))
plt.scatter(x_pca[:,0], x_pca[:,1], c = cancer['target'],cmap= 'plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns = cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df_comp, cmap= 'plasma')                               # the more warm color means larger the number then it is more correlated to its feature or column.
```
