#  K Means Clustering

Clustering a grouping together similar unlabelled data points. It is an Unsupervised Learning. 

In this,
First we will use the K value (means the no of Clusters)
then assign each point to a cluster.
Until cluster stop changing, Repeat the following:
    For each cluster, compute the cluster centroid by taking the mean vector of the points in cluster.
    Assign each data point to a cluster for which the cenroid is the closest.

For choosing the K Value, we have to use the ELBOW Method.
First, compute the sum of squared error for some value of k.
SSE means the sum of squared distance between each member of its cluster and its Centroid.

For large value of K, SSE (error) is less.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs( n_samples=200,n_features=2,centers=5,cluster_std=1.8,random_state = 101)
```

```python
# data is a tuple which has 2d numpy array
# data[0] is a data and ( data[1] is labels, it is an artificial data ) it is unsupervised which is unlabelled.
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c = data[1] , cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 5)                                 # we took 5 because we know as it is an artificial data. If we use real data then we have to assume it.
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_                                         # Predicted centroids
```

```python
kmeans.labels_                                # Predicted labels.
```

```python
plt.scatter(data[0][:,0], data[0][:,1],c = kmeans.labels_ , cmap = 'rainbow')

#If we change n_clusters=5 to any other number (where we had define the model) then we will get that no of clusters in prediction
```
