# K Means Clustering using Sklearn

Clustering is an unsupervised learning, <br>
Clustering is different from Classification. <br>
For classification, we need class. <br>
but for clustering, we need no class initially. we are just clustering up the things which has no label associated with it.

__KNN vs K Means Clustering__ <br>
KNN has labelled classes. A new input is set to the class to which that input point is nearest to. (Supervised Learning)<br>
K means Clustering do not have labelled classes. They form cluster according to the similarity and then they form classes. Means provide y or outputs to the data. (Unsupervised Learning)

```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
X, _ = make_blobs(centers=3, random_state=42)
```

```python
sns.scatterplot(X[:,0], X[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(X)
```

```python
y_gen = model.labels_
```

```python
y_pred = model.predict(X)
```

```python
sns.scatterplot(X[:,0], X[:,1], hue = y_gen)
```

```python
model.cluster_centers_
```

```python
sns.scatterplot(X[:,0], X[:,1], hue = y_gen)

for center in model.cluster_centers_:
    plt.scatter(center[0], center[1], s = 60)
```

```python
model2 = KMeans(n_clusters=2)
model2.fit(X)
y_gen2 = model2.labels_



sns.scatterplot(X[:,0], X[:,1], hue = y_gen2)

for center in model2.cluster_centers_:
    plt.scatter(center[0], center[1], s = 60)
```
